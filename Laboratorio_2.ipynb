# ***Ensayo - Git***

Git es un Version Control System, que sirve para dar seguimiento a las versiones del cdigo relacionado a un archivo o   proyecto. Los sistemas de control de versiones facilitan el seguimiento a proyectos donde colaboran muchas personas.

Git se ha convertido en el estndar de los sistemas de control de versiones, por su parte Git hace una captura del archivo y si hubo un cambio, puede generar detalles de:

        - Quien edit el archivo
        - Cuando lo hizo
        - Que linea de cdigo cambi
